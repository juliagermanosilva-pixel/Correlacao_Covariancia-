```python
import numpy as np
alturas = np.array ([160,170,175,163,180])
#peak-to-peak
print(alturas)
amplitude = np.ptp(alturas)
print(amplitude)
#amplitude = maior valor - menor valor
```

```python
#desvio padrãoo
media_alturas = np.mean(alturas)
print(media_alturas)
#calculo de desvio padrão
desvio_padrao = np.std(alturas, ddof=1)
print(desvio_padrao)
```

```python
#covariância positivo
#quando duas coisas crescem ou decrescem ao mesmo tempo

#covariância negativa
#quando uma variável cresce e a outra decresce

#covariância zero
#quando nada muda
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
dados_equipamento = {
    'consumo': [10,20,30,40,50,60,70,80],
    'temperatura': [25,32,45,48,60,72,85,90]
}
#transformar esses dados para o pandas ler
df_dados = pd.DataFrame(dados_equipamento)
#vamos fazer o calculo de covariância
#criar uma base para a covariancia
cov_matrix = df_dados.cov()
print(f'a covariância foi {cov_matrix.iloc[0,1]}')
#corr - função para correlação
correlacao_matrix = df_dados.corr()
print(correlacao_matrix)
#gerar um grafico para essas correlações
sns.scatterplot(x='consumo', y='temperatura', data=df_dados)
plt.title('Gráfico de Correlação')
plt.show()
```

## Visualizações Adicionais

Vamos adicionar mais algumas visualizações para entender melhor a distribuição dos dados de consumo e temperatura.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de Barras para Consumo
plt.figure(figsize=(10, 6))
sns.barplot(x='consumo', y='temperatura', data=df_dados, palette='viridis')
plt.title('Consumo vs. Temperatura (Gráfico de Barras)')
plt.xlabel('Consumo')
plt.ylabel('Temperatura')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Gráfico de Pizza para Distribuição de Consumo

# Criar categorias de consumo para o gráfico de pizza
bins = [0, 30, 60, 90] # Definir os limites dos bins
labels = ['Baixo', 'Médio', 'Alto'] # Rótulos para as categorias

df_dados['consumo_categoria'] = pd.cut(df_dados['consumo'], bins=bins, labels=labels, right=False)

# Contar a frequência de cada categoria
consumo_counts = df_dados['consumo_categoria'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(consumo_counts, labels=consumo_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribuição de Categorias de Consumo')
plt.axis('equal') # Garante que o gráfico de pizza seja um círculo.
plt.show()
```
